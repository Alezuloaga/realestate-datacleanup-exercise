Imprime ambos promedios y escribe una conclusión sobre ellos.

```python
# TODO
media_valdemorillo = ds[ds['level5'] == "Valdemorillo"]['price'].mean()
media_galapagar = ds[ds['level5'] == "Galapagar"]['price'].mean()

print(f"Precio promedio en Valdemorillo: {media_valdemorillo:.2f} USD")
print(f"Precio promedio en Galapagar: {media_galapagar:.2f} USD")

if media_valdemorillo == media_galapagar:
    print("Conclusión: Los precios promedio son exactamente iguales.")
else:
    diferencia = abs(media_valdemorillo - media_galapagar)
    print(f"Conclusión: No son iguales. Hay una diferencia de {diferencia:.2f} USD.")
```

#### Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

Para el método de normalización, puedes usar el que consideres adecuado, no hay una única respuesta correcta para esta pregunta. Imprime el gráfico y escribe en la celda de Markdown un breve análisis sobre el gráfico.

Pista: Puedes ayudarte revisando la demostración multihist de Matplotlib.

```python
# TODO
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

for poblacion in ciudades_sur:
    precios = ds_sur[ds_sur['level5'] == poblacion]['price']
    
    precios_norm = (precios - precios.min()) / (precios.max() - precios.min())
    
    plt.hist(precios_norm, bins=15, alpha=0.7, label=poblacion, histtype='step', linewidth=2)

plt.title('Distribución Normalizada de Precios por Población')
plt.legend()
plt.show()
```

#### Ejercicio 21. Realiza un trazado de las coordenadas (columnas latitud y longitud) del cinturón sur de Madrid por color de cada población (debes usar el subconjunto obtenido del Ejercicio 14) (★★★★)

Ejecuta la siguiente celda y luego comienza a codear en la siguiente. Debes implementar un código simple que transforme las columnas de coordenadas en un diccionario de Python (agrega más información si es necesario) y agrégala al mapa.

```python
from ipyleaflet import Map, basemaps

# Mapa centrado en Madrid
map = Map(center = (40.35, -3.8), zoom = 11.5, min_zoom = 1, max_zoom = 20, 
    basemap=basemaps.OpenStreetMap.Mapnik)
map
```

```python
## Aquí: traza la coordenadas de los estados

from ipyleaflet import Marker, MarkerCluster

colores_poblacion = {
    "Fuenlabrada": "red",
    "Leganés": "blue",
    "Getafe": "green",
    "Alcorcón": "orange"
}

## PON TU CÓDIGO AQUÍ:
marcadores = []

for index, fila in ds_sur.iterrows():
    
    if pd.notnull(fila['latitude']) and pd.notnull(fila['longitude']):
        marcador = Marker(
            location=(fila['latitude'], fila['longitude']),
            draggable=False,
            title=f"{fila['level5']} - {fila['price']} USD"
        )
        marcadores.append(marcador)
agrupacion_puntos = MarkerCluster(markers=marcadores)
map.add_layer(agrupacion_puntos)

map
```
